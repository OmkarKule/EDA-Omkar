```python
print(df['Embarked'].value_counts())
```

```python
plt.figure(figsize=(7,9))
#print(df['Embarked'].value_counts
plt.pie(df['Embarked'].value_counts(),
        labels=df['Embarked'].unique()[:3],
        autopct='%1.2f%%',
        startangle=90,
        explode=[0,0,0.2]
       )
plt.legend()
```
